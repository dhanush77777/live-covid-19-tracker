```python
import pandas as pd
df=pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
df
```

```python
df=df.iloc[:-1,1:]
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## tracking the positive cases in all states

```python
plt.figure(figsize=(9,8))
sns.barplot(y=df['state_name'],x=df['positive']) 
plt.xlabel('number of positive cases')
plt.ylabel('state name')
plt.title('covid cases')
plt.show()
```

```python
positive=df.groupby('state_name')['positive'].mean().reset_index()
positive.head()
```

## tracking the active cases in all states

```python
plt.figure(figsize=(9,8))
sns.barplot(y=df['state_name'],x=df['active'],color='b') 
plt.xlabel('number of active cases')
plt.ylabel('state name')
plt.title('covid cases')
plt.show()
```

```python
active=df.groupby('state_name')['active'].mean().reset_index()
active.head()
```

## tracking the death cases in all states

```python
plt.figure(figsize=(9,8))
sns.barplot(y=df['state_name'],x=df['death'],color='r') 
plt.xlabel('number of death cases')
plt.title('covid cases')
plt.show()
```

```python
death=df.groupby('state_name')['death'].mean().reset_index()
death.head()
```
